```python
import pandas as pd
from sqlalchemy import create_engine
```

## Extract CSVs into DataFrames

```python
csv_file = "Project youtube/JPvideos.csv."
JPvideos_df = pd.read_csv(csv_file, encoding="ISO-8859-1")
JPvideos_df.head()
```

```python
csv_file = "Project youtube/DEvideos.csv."
DEvideos_df = pd.read_csv(csv_file)
DEvideos_df.head()
```

```python
csv_file = "Project youtube/USvideos.csv."
USvideos_df = pd.read_csv(csv_file)
USvideos_df.head()
```

```python
US_json = "Project youtube/US_category_id.json"
US_json = pd.read_json(US_json)
for video in US_json['items']:
    print('-'*30)
    print(video['kind'])
    print(video['id'])
    print(video['snippet']['channelId'])
    print(video['snippet']['title'])
    
video.keys()
```

## Data Transformation 

```python
# Create a filtered dataframe from specific columns
new_JPvideos_df = JPvideos_df[["trending_date", "title", "channel_title", "views", "likes", "dislikes", "comment_count"]].copy()
new_JPvideos_df.head()
```

```python
new_DEvideos_df = DEvideos_df[["trending_date", "title", "channel_title", "views", "likes", "dislikes", "comment_count"]].copy()
new_DEvideos_df.head()
```

```python
new_USvideos_df = USvideos_df[["trending_date", "title", "channel_title", "views", "likes", "dislikes", "comment_count"]].copy()
new_USvideos_df.head()
```

## Create database connection

```python
import pymysql
pymysql.install_as_MySQLdb()

rds_connection_string = "root:Mummys12$@127.0.0.1/youtube_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

## Load DataFrames into database

```python
new_USvideos_df.to_sql(name='usvideos', con=engine, if_exists='append', index=False)
```

```python
new_DEvideos_df.to_sql(name='devideos', con=engine, if_exists='append', index=False)
```
